```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('First Graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('First Part of the Graph')
plt.subplot(1,2,2)
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('Second Part of the Graph')
plt.plot(y,x)
plt.tight_layout
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.sub_x
```

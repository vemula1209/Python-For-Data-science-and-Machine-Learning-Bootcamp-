```python
y = customer['Yearly Amount Spent']
```

```python
X = customer[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
customer.head()
```

```python
lm.coef_
```

```python
prediction = lm.predict(X_test)
```

```python
plt.scatter(y_test,prediction)
plt.xlabel('Y Test(True values)')
plt.ylabel('Predected values')
```

```python
plt.xlabel('Y Test(True values)')
```

```python
plt.ylabel('Predected values')
```

```python
plt.scatter(y_test,prediction)
```

```python
from sklearn import metrics
```

```python
print('MAE',metrics.mean_absolute_error(y_test,prediction))
print('MSE',metrics.mean_squared_error(y_test,prediction))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,prediction)))
print('Variance',metrics.explained_variance_score(y_test,prediction))
```

```python
sns.distplot((y_test-prediction),bins=50)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python

x= np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.title('First Figure')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('X axes')
axes.set_ylabel('Y axes')
axes.set_title('Object Oriented Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes1.plot(x,y)
axes1.set_xlabel('X axes')
axes1.set_ylabel('Y axes')
axes1.set_title('First Graph')
axes2 = fig.add_axes([.2,.5,.4,.3])
axes2.plot(y,x)
axes2.set_xlabel('X axes')
axes2.set_ylabel('Y axes')
axes2.set_title('SECONd Graph')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('X axes')
axes[0].set_xlabel('Y axes')
axes[0].set_title('First Graph')
axes[1].plot(y,x)
axes[1].set_xlabel('X axes')
axes[1].set_xlabel('Y axes')
axes[1].set_title('Second Graph')
plt.tight_layout()
#axes[0].set_xlabel('X axes')
#axes[0].set_ylabel('Y axes')
#axse[0].set_title('First Graph')

#axes[1].plot(Y,X)
#axes[1].set_xlabel('X axes')
#axes[1].set_ylabel('Y axes')
#axse[1].set_title('First Graph')
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('New Grah')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('First Graph')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Second Graph')
```

```python
fig = plt.figure()
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
axis.set_xlabel('X Axis')
axis.set_ylabel('Y Axis')
axis.set_title('Object Graph')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('First Graph')

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y)
axes2.set_xlabel('X axis')
axes2.set_ylabel('Y axis')
axes2.set_title('Second Graph')
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y =  x**2
```

```python
plt.plot(x,y)
plt.xlabel('X axies')
plt.ylabel('Y axies')
plt.title('First Graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'Red')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.title('First Graph')
plt.subplot(1,2,2)
plt.plot(y,x,'Green')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.title('Second Graph')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)           
axes.set_xlabel('X axes')
axes.set_ylabel('Y Axes')
axes.set_title('First Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'green')
axes1.set_xlabel('X axes')
axes1.set_ylabel('Y axes')
axes1.set_title('First Graph')
axes2 = fig.add_axes([0.2,0.4,0.3,0.4])
axes2.plot(y,x,'red')
axes2.set_xlabel('X axes')
axes2.set_ylabel('Y axes')
axes2.set_title('Second Graph')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('X axes')
axes[0].set_ylabel('Y axes')
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
axes[1].set_xlabel('X axes')
axes[1].set_ylabel('Y axes')
```
